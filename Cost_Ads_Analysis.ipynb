**Analisis Pengeluaran**

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import plotly.express as px

# Load data from Excel file
file_path = '/content/Ads_Campaign_Report.xlsx'  # Ganti dengan path sesuai lokasi file Anda
df = pd.read_excel(file_path)

# Pilih kolom yang diperlukan
df = df[['Tanggal', 'Pengeluaran']]

# Ubah nama kolom agar sesuai
df.columns = ['ds', 'y']

# Mengubah kolom 'Tanggal' ke tipe datetime
df['ds'] = pd.to_datetime(df['ds'])

# Mengurutkan data berdasarkan tanggal
df.sort_values('ds', inplace=True)

# Menggunakan ARIMA untuk forecasting
model = ARIMA(df['y'], order=(1, 1, 1))  # Ganti parameter order sesuai kebutuhan
result = model.fit()

# Forecasting 3 bulan ke depan
forecast = result.get_forecast(steps=3*30)  # 3 bulan * 30 hari/bulan
forecast_index = pd.date_range(start=df['ds'].max(), periods=3*30+1, freq='D')[1:]  # Index untuk hasil forecast

# Membuat DataFrame untuk hasil forecast
forecast_df = pd.DataFrame({'ds': forecast_index, 'yhat': forecast.predicted_mean})

# Membuat grafik garis interaktif menggunakan Plotly Express
fig = px.line(df, x='ds', y='y', title='ARIMA Forecast Pengeluaran untuk 3 Bulan ke Depan')
fig.add_scatter(x=forecast_df['ds'], y=forecast_df['yhat'], mode='lines', name='ARIMA Forecast', line=dict(dash='dash', color='orange'))
fig.update_xaxes(title_text='Tanggal')
fig.update_yaxes(title_text='Pengeluaran')
fig.show()
```
