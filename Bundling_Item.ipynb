```python
import pandas as pd

# Membaca data dari file CSV
file_path = 'Item All Merchant.csv'
df = pd.read_csv(file_path)

# Langkah 1: Pisahkan item makanan dan minuman berdasarkan category_name
food_items = df[df['category_name'].str.contains('FOOD', case=False, na=False)]
drink_items = df[df['category_name'].str.contains('DRINK', case=False, na=False)]

# Langkah 2: Pilih item makanan dan minuman yang sesuai
food_selection = food_items['item_name'].unique()
drink_selection = drink_items['item_name'].unique()

# Langkah 3: Kombinasikan item makanan dan minuman dalam bundling
bundles = []
for food_item in food_selection:
    for drink_item in drink_selection:
        bundles.append([food_item, drink_item])

# Menampilkan bundling yang telah dibuat
print("Bundling Makanan dan Minuman:")
for bundle in bundles:
    print("Bundling:", ', '.join(bundle))
```
